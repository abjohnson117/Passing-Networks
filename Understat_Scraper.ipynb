```python
def in_degree(G: nx.DiGraph(), title: str, degree=True):
    """Find the in-degree distribution of a passing network"""
    # This will get the in-degree distribution
    if degree:
        A = nx.adjacency_matrix(G).todense().T
        ones = np.ones(len(A))
        deg_seq = A@ones
        deg_seq = np.squeeze(np.asarray(deg_seq))

        degree_count = Counter(deg_seq)
        deg, cnt = zip(*degree_count.items())
        deg, cnt = list(deg), list(cnt)
        for i in range(len(cnt)):
            cnt[i] = cnt[i] / len(list(G.nodes))

        plt.bar(deg, cnt)
        plt.xlabel("Degree")
        plt.ylabel("Frequency")
#         plt.loglog()
        plt.title(title)
    # This will get the out-degree distribution
    else:
        A = nx.adjacency_matrix(G).todense()
        ones = np.ones(len(A))
        deg_seq = A@ones
        deg_seq = np.squeeze(np.asarray(deg_seq))

        degree_count = Counter(deg_seq)
        deg, cnt = zip(*degree_count.items())
        deg, cnt = list(deg), list(cnt)
        for i in range(len(cnt)):
            cnt[i] = cnt[i] / len(list(G.nodes))

        plt.bar(deg, cnt)
        plt.xlabel("Degree")
        plt.ylabel("Frequency")
#         plt.loglog()
        plt.title(title)
```

```python
def pass_centrality(G: nx.DiGraph()):
    """Calculate network centrality measure devised by Thomas Grund, as detailed in Soccermatics"""
    A = nx.adjacency_matrix(G).todense()
    ones = np.ones(len(A))
    max_recipients = A@ones
    max_recipient = np.amax(max_recipients)
    max_recipients = np.squeeze(np.asarray(max_recipients))
    
    arg = np.argmax(max_recipients)
    pass_differences = []
    for i in max_recipients:
        pass_differences.append(max_recipient - i)
    pass_differences = np.array(pass_differences)
    
    passes_made = np.sum(max_recipients)
    pass_diff_sum = np.sum(pass_differences) # was sum
    pass_centrality = pass_diff_sum / (10 * passes_made)
    
    most_received = list(G.nodes())[arg]
    
    return pass_centrality, most_received
```

```python
def get_pass_network(match_id: int, team: str, color_1: str, color_2: str, save_image=False):
    """This function cleans the StatsBomb data and outputs a nice graphical image on a soccer pitch
    and it also creates the NetworkX graph object"""
    # First get lineups
    madrid_lineups = sb.lineups(match_id=match_id)[team]
    madrid_lineups = madrid_lineups[["player_name", "jersey_number"]]
    madrid_lineups.columns = ["player", "playerid"]
    
    # Get the actual match-level data and start cleaning
    madrid = sb.events(match_id=match_id)
    madrid["pass_outcome"] = madrid["pass_outcome"].fillna("Successful")
    madrid = madrid.iloc[6:, :]
    subs = madrid.loc[madrid["type"] == "Substitution"].loc[madrid["team"] == team].iloc[0]["index"]
    madrid = madrid.loc[madrid["type"] == "Pass"]
    
    madrid[["x", "y"]] = pd.DataFrame(madrid["location"].to_list(), index=madrid.index)
    madrid[["endX", "endY"]] = pd.DataFrame(madrid["pass_end_location"].to_list(), index=madrid.index)
    madrid = madrid.loc[madrid["team"] == team]
    madrid = madrid[["timestamp", "index", "team", "type", "player", "pass_outcome", "x", "y", "endX", "endY"]]
    madrid = madrid.merge(madrid.merge(madrid_lineups, on="player", sort=False))
    
    madrid = madrid.drop(columns=["player"])
    madrid["passer"] = madrid["playerid"]
    madrid["recipient"] = madrid["playerid"].shift(-1)

    successful = madrid[madrid["pass_outcome"] == "Successful"]
    successful["recipient"] = successful["recipient"].fillna(8.0)
    
    successful = successful.loc[successful["index"] < subs]
    
    pas = pd.to_numeric(successful["passer"], downcast="integer")
    rec = pd.to_numeric(successful["recipient"], downcast="integer")

    successful["passer"] = pas
    successful["recipient"] = rec
    
    average_locations = successful.groupby("passer").agg({"x": ["mean"], "y": ["mean", "count"]})
    average_locations.columns = ["x", "y", "count"]
    
    pass_between = successful.groupby(["passer", "recipient"]).count().reset_index()
    pass_between.rename({"id": "pass_count"}, axis=1, inplace=True)

    pass_between = pass_between.merge(average_locations, left_on="passer", right_index=True)
    pass_between = pass_between.merge(average_locations, left_on="recipient", right_index=True, suffixes=["","_end"])

    pass_between = pass_between[["passer", "recipient", "x_y", "y_y", "count", "x", "y", "count_end"]]
    
    # Plot the Pitch
    pitch = Pitch(pitch_type='statsbomb', pitch_color='#22312b', line_color='#c7d5cc')
    fig, ax = pitch.draw(figsize=(16, 11), constrained_layout=False, tight_layout=True)
    fig.set_facecolor('#22312b')

    arrows = pitch.arrows(1.2*pass_between["x_y"], .8*pass_between["y_y"], 1.2*pass_between["x"], 
                          .8*pass_between["y"],
                         ax=ax, width=3, headwidth=3, color=color_1, zorder=1, alpha=0.5)

    nodes = pitch.scatter(1.2*average_locations["x"], .8*average_locations["y"], s=300, color=color_2,
                         edgecolors="black", linewidth=2.5, alpha=1, zorder=1, ax=ax)
    
    for index, row in average_locations.iterrows():
        pitch.annotate(row.name, xy=(1.2*row.x, .9*row.y), c='gold', va='center',
                   ha='center', size=18, weight='bold', ax=ax)
    
    if save_image:
        fig.savefig(team + ".jpeg")
    
    # Get the actual network structure
    nodes = pass_between["passer"].unique()
    edges = list(zip(pass_between["passer"].to_list(), pass_between["recipient"].to_list()))
    count_connections = list(zip(successful["passer"].to_list(), successful["recipient"].to_list()))
    count_connections = dict(Counter(count_connections))
    pre_pos = list(zip(average_locations["x"].to_list(), average_locations["y"].to_list()))
    pos = {}
    for i in range(len(pre_pos)):
        pos[nodes[i]] = pre_pos[i]
        
    edge_list = []
    for edge in edges:
        edge_list.append((edge[0], edge[1], count_connections[edge]))
        
    G = nx.DiGraph()
    G.add_nodes_from(nodes)
    G.add_weighted_edges_from(edge_list)
    
    return G, pos
```
